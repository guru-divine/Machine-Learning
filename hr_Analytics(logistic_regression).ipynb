```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("HR.csv")
df.head()
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
dummy_sal = pd.get_dummies(df.salary)
dummy_dep = pd.get_dummies(df.Department)
total = pd.concat([df,dummy_sal,dummy_dep], axis='columns')
total.head()
```

```python
# total = total.drop(["Department", "salary"], axis='columns')
total.head()
```

```python
X_data = total.drop(["left"], axis='columns')
X_data
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_data, total.left, train_size=0.8)
```

```python
X_test.head()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
