```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
```

```python
df = pd.read_csv("carprices.csv")
df
```

```python
dummies = pd.get_dummies(df.loc[:,"Car Model"])
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(["Car Model", "Mercedez Benz C class"], axis='columns')
final
```

```python
X = final.drop("Sell Price($)", axis = "columns")
X
```

```python
y = final.loc[:,"Sell Price($)"]
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
model.predict([[50000,2,0,0]])
```

```python
model.predict([[65000, 5,1,0]])
```

```python
import joblib
joblib.dump(model, "car_price_predictor")
```

```python
model.score(X,y)
```
